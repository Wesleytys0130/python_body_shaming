## 隨機取樣
隨機抽取樣本與標記好的資料混合，之後再訓練模型。

```python
import pandas as pd
import re
random_df = pd.read_csv("/Users/guanlizhe/Desktop/nccu/NLP/body_shaming/ptt_data/Gossiping/index10-11.csv", index_col=[0],low_memory=False)
random_df = random_df.sample(n=7225,axis=0)
# other_df.to_csv("other_df.csv",mode='w', encoding='utf-8')
# random_df
```

### 取出的樣本做一樣的標記

```python
random_df['year'] = pd.to_numeric(random_df['year'],errors='coerce')
random_df['year'] = random_df['year'].fillna(value= 2022.0)
random_df['year'] = random_df['year'].astype(int)

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder() 
features_tag = random_df[['tag']].apply(LabelEncoder().fit_transform)
features_tag
random_df.insert(2,"tag_num",features_tag,True)

list0 = [0 for i in range(0,len(random_df))]
random_df.insert(6,"label",list0,True)

similar_animal = []
for comment in random_df['content']:
    comment = str(comment)
    found = re.findall('豬|龍|猴|狗|鳥|蟲|象|畜|馬|河豚|鮪魚|蛞蝓', comment)
    count = re.findall('歐米龍|萊豬|印象|不死鳥', comment)
    if len(found) > 0 and len(count) == 0:
        similar_animal.append(1)
    else:
        similar_animal.append(0)
random_df['similar_animal'] = similar_animal

similar_object = []
for comment in random_df['content']:
    comment = str(comment)
    found = re.findall('盤子|水桶|竹竿|紙片|神棍|甜不辣|難民|骨|骷髏|臘腸|塑膠|火腿|騎士|奴隸|划龍舟|瓦斯', comment)
    if len(found) > 0:
        similar_object.append(1)
    else:
        similar_object.append(0)
random_df['similar_object'] = similar_object

body = []
for comment in random_df['content']:
    comment = str(comment)
    found = re.findall('臉|腿|腳|手|肚|屁|腰', comment)
    count = re.findall('打腫臉|柯屁|腰帶', comment)
    if len(found) > 0 and len(count) == 0:
        body.append(1)
    else:
        body.append(0)
random_df['body'] = body

direct_adj = []
for comment in random_df['content']:
    comment = str(comment)
    found = re.findall('醜|窮|胖|肥|矮|短|粗|臭|噁|垂|厚|禿|油|老|扁|鬼|髒|魯|殘|肉|腫|低能|不正|假掰|心機|心地惡毒', comment)
    count = re.findall('安迪魯伊茲|胖虎|胖卡|金小胖|矮油|矮堅', comment)
    if len(found) > 0 and len(count) == 0:
        direct_adj.append(1)
    else:
        direct_adj.append(0)
random_df['direct_adj'] = direct_adj

verb = []
for comment in random_df['content']:
    comment = str(comment)
    found = re.findall('滾|跑|減肥|閉嘴|嘔|騷擾', comment)
    if len(found) > 0:
        verb.append(1)
    else:
        verb.append(0)
random_df['verb'] = verb

slurs = []
for comment in random_df['content']:
    comment = str(comment)
    found = re.findall('垃圾|廢物|馬的|媽的|操|幹|幹你媽|幹你娘|死|要命|跩|賤', comment)
    count = re.findall('幹嘛|幹什麼|幹麼', comment)
    if len(found)>0 and len(count)==0:
        slurs.append(1)
    else:
        slurs.append(0)
random_df['slur'] = slurs

sex = []
for comment in random_df['content']:
    comment = str(comment)
    found = re.findall('屌|肛|自慰|姦|包莖|陰道|鮑|處男|處女|聖人模式|醜囊|喝茶|嬯|台女|墮胎|幹|A片|a片|破麻|破處|肏|督|抽插|陰唇|海鮮|晨勃|勃起|雞雞|做愛|貞操|強暴|洨|肉便器|婊|妓女|撞起來|能硬|打手槍|西斯|趴起來|ㄐㄐ', comment)
    count = re.findall('洨胖|洨弟|洨妹', comment)
    if len(found) > 0 and len(count) == 0:
        sex.append(1)
    else:
        sex.append(0)
random_df['sex'] = sex

lgbt = []
for comment in random_df['content']:
    comment = str(comment)
    found = re.findall('甲|熊|女T|娘娘腔|變性|跨性別|gay|Gay|GAY', comment)
    if len(found) > 0:
        lgbt.append(1)
    else:
        lgbt.append(0)
random_df['lgbt'] = lgbt

status = []
for comment in random_df['content']:
    comment = str(comment)
    found = re.findall('8.?9|/+9|黑道|結婚|娶|大媽|阿姨|阿桑|阿罵|婆媽|歐巴桑|大叔', comment)
    count = re.findall('未成年結婚', comment)
    if len(found) > 0 and len(count) == 0:
        status.append(1)
    else:
        status.append(0)
random_df['status'] = status

# list0 = [0 for i in range(0,len(random_df))]
random_df.insert(16,"label_ver2",list0,True)

random_df
```

## 合併資料
把取出的樣本處理完，和原始資料做合併，合併後把順序打亂再重新排序index。

```python
df = pd.read_csv("/Users/guanlizhe/Desktop/nccu/NLP/body_shaming/all_label.csv", index_col=[0],low_memory=False)
df = df.drop(columns=['label_ver1'])
df = df.sample(n=7225,axis=0)
random_df = pd.concat([random_df,df])
random_df = random_df.sample(n=len(random_df),axis=0)
random_df.reset_index(drop=True, inplace=True)

random_df
```

```python
random_df.to_csv("random_df.csv",mode='w', encoding='utf-8')
```
